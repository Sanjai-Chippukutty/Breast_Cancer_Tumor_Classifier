##  Visualizing Class Distribution

After loading and inspecting the dataset, we explore the distribution of our **target variable** — `benign_0__mal_1`. This binary column indicates whether the tumor is **benign (0)** or **malignant (1)**.

Understanding class distribution is essential in classification problems. An imbalanced dataset (e.g., many more benign than malignant cases) may bias the model during training and affect predictive performance.

---

###  Target Column: `benign_0__mal_1`

- **0** = Benign Tumor  
- **1** = Malignant Tumor

We use **Seaborn’s `countplot()`** to visualize the frequency of each class.




```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset again
file_path = "C:/Users/sanja/1. Breast_Cancer_Tumor_Classifier/1.Breast_Cancer_Tumor_Classifier/data/raw/cancer_classification.csv"
df = pd.read_csv(file_path)

# Set the target column
target_column = 'benign_0__mal_1'

# Plot class distribution
plt.figure(figsize=(6,4))
sns.countplot(data=df, x=target_column, palette=['green', 'red'])
plt.title('Class Distribution')
plt.xlabel('Tumor Type (0 = Benign, 1 = Malignant)')
plt.ylabel('Count')
plt.xticks([0, 1], ['Benign (0)', 'Malignant (1)'])
plt.show()
```

##  Correlation Analysis – Heatmap

To understand the relationships between different numerical features in our dataset, we compute and visualize the **Pearson correlation matrix**.

This heatmap helps us:
- Identify **strongly correlated features** (positively or negatively)
- Detect **redundant features** (high correlation with each other)
- Spot **useful predictors** for our target variable

---

###  What is Correlation?

Correlation is a statistical measure that indicates the extent to which two variables fluctuate together.

- **+1** → Strong positive correlation  
- **0** → No correlation  
- **–1** → Strong negative correlation

---

###  Plotting the Correlation Heatmap

We use `pandas.corr()` to compute the correlation matrix and visualize it using Seaborn's `heatmap()`.




```python
# Correlation heatmap (top 15 for clarity)
plt.figure(figsize=(14, 12))
correlation_matrix = df.corr()

# Draw the heatmap
sns.heatmap(correlation_matrix, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```
